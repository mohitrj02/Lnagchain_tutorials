### Agent with Memory

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver # saves the curent session within the memory of local machine and reuse that as context. 
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model = "gemini-2.5-flash-lite")

# agent with a property that is InMemorySaver. Basically, now our agent has a memory.
agent = create_agent(
    model = model,
    checkpointer = InMemorySaver()
)
```

Before any conversation we need to create a **Thread** for maintaining the session.

```python
# Run with thread id 1
config = {"configurable":{"thread_id":"test1"}}
```

According to langgraph, whenever we use InMemorySaver so their we have to specify a key i.e. configurable with a unique thread id to specify a single user.

```python
from langchain_core.messages import HumanMessage, SystemMessage

response = agent.invoke(
    {"messages":[HumanMessage(content = "Hi! My name is Mohit!")]},
    config = config
)
response

```

```python
agent.invoke({"messages":[HumanMessage(content="Hi! What is my name?")]},config=config)
```

```python
# Run with thread id 2
config1 = {"configurable":{"thread_id":"test2"}}

from langchain_core.messages import HumanMessage, SystemMessage

response = agent.invoke(
    {"messages":[HumanMessage(content = "Hi! My name is Alice!")]},
    config = config1
)
response
```
