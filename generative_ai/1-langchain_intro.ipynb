### LLM Model Integration With Google Gemini

Import the API_KEY

```python
os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
```

Call the Model

```python
from langchain.chat_models import init_chat_model
model=init_chat_model(
    model="google_genai:gemini-2.5-flash-lite"
)
response=model.invoke("Write me an essay about Artificial Intelligence")
print(response.content)
```

### Streaming and Batch

#### Streaming
Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses. Calling stream() returns an iterator that yeilds output chunks as they are produced. You can use a loop to process each chunk in real-time:

Normal way:

```python
model.invoke("Write me a 500 words paragraph on AI")
```

Streaming way:

```python
model.stream("Write me a 500 words paragraph on AI")
```

```python
for chunk in model.stream("Write me a 500 words paragraph on AI"):
    print(chunk.text, end="", flush=True)
```

Example 2:

```python
model.invoke("Why do parrots have colorful feathers?")
```

```python
for chunk in model.stream("Why do parrots have colorful feathers?"):
    print(chunk.text, end="", flush=True)
```

#### Batch
Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel:

```python
responses = model.batch(
    [
        "Why do parrots have colorful feathers?",
        "How do airplace fly?",
        "What is quantum computing"
    ]
)
for response in responses:
    print(response)
```

Limit the batch size:

```python
model.batch(
    [
        "Why do parrots have colorful feathers?",
        "How do airplace fly?",
        "What is quantum computing"
    ],
    config = {
        'max_concurrency': 5, # Limit to 5 parallel calls
    }
)
```
